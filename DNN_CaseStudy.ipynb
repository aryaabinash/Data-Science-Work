# Case No 1 : Perceptron Model

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits
```

```python
digits['data']
```

```python
digits.target
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.linear_model import Perceptron
```

```python
x = digits.data
y = digits.target
```

```python
pd.DataFrame(x).shape
```

```python
pd.DataFrame(y).value_counts()
```

```python
pd.DataFrame(y).nunique()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Building perceptron model
pp = Perceptron()
pp.fit(x_train, y_train)
print("What is the Testing Accuracy")
print(pp.score(x_test, y_test))
```

```python
print("What is the Training Accuracy")
print(pp.score(x_train, y_train))
```

```python
predicted = pp.predict(x_test)
```

```python
print('\n confusion_matrix')
print(confusion_matrix(y_test, predicted))
```

```python
print('\n classification_report')
print(classification_report(y_test, predicted))
```

# Case number 2 : MultiLayer Perceptron Model - Regression Problem

```python
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
dataset = pd.read_csv('HousingData.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# Handling missing value by using KNN imputation 
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer()
dataset1 = pd.DataFrame(imputer.fit_transform(dataset), columns = dataset.columns)
```

```python
dataset1.isnull().sum()
```

```python
dataset1.info()
```

```python
x = dataset1.iloc[:,:-1]
y = dataset1.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# Feature Scaling - Mandatory step
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
scaled_x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.25, random_state=42)
```

# MultiLayer Perceptron Model - Regression

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
# Initialize the Sequential model
model = Sequential()
# First fully connected Hidden Layer 
model.add(Dense(32, activation='relu', input_dim=13))
# Second fully connected Hidden Layer 
model.add(Dense(16, activation='relu'))
# output
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=32)
# Gradient Descent optimization (BGD, SGD, MBGD)
```
