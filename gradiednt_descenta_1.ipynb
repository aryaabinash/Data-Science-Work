```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m = 0
    c = 0
    n = len(x)
    iterations = 1000
    learning_rate = 0.01
    # Lists to store history for plotting
    cost_history = []
    iteration_history = []
    for i in range(iterations):
        y_predicted = m * x + c
        # Calculating derivatives
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        dm = (-2/n) * sum(x * (y - y_predicted))
        dc = (-2/n) * sum(y - y_predicted)

        # Updating parameters
        m = m - learning_rate * dm
        c = c - learning_rate * dc
        
        # Save values to history
        cost_history.append(cost)
        iteration_history.append(i + 1)
        # print("slope = {}, intercept = {}, iter={}, cost = {}".format(m,c,i+1,cost))
        # Optional: print every 100 iterations to avoid cluttering the console
        if (i + 1) % 100 == 0:
            print(f"Iteration {i+1}: Cost {cost}")

    return iteration_history, cost_history
        
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
```

```python
# Run the gradient descent
iters, costs = gradient_descent(x, y)

# PLot the line between iteration vs cost
plt.plot(iters, costs, color='red',linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Cost (MSE)')
plt.title("Gradient Desent: Iteration vs Cost")
plt.grid(True)
plt.savefig('iteration_vs_cost.png')
```

```python
# Plot the choosen line(Best fit line)

import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m = 0
    c = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.05
    
    for i in range(iterations):
        y_predicted = m * x + c
        # Calculating derivatives
        dm = (-2/n) * sum(x * (y - y_predicted))
        dc = (-2/n) * sum(y - y_predicted)
        
        # Updating parameters
        m = m - learning_rate * dm
        c = c - learning_rate * dc
        
    return m, c

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Get final slope and intercept after 1000 iterations
m_final, c_final = gradient_descent(x, y)

# Calculate the points for the best fit line
y_final_predicted = m_final * x + c_final

# Plotting
plt.scatter(x, y, color='red', marker='+', label='Original Data') # Scatter plot of data
plt.plot(x, y_final_predicted, color='blue', label='Best Fit Line') # Line plot of prediction
plt.xlabel('x')
plt.ylabel('y')
plt.title('Best Fit Line using Gradient Descent')
plt.legend()
plt.grid(True)
plt.savefig('best_fit_line.png')

print(f"Final slope (m): {m_final}")
print(f"Final intercept (c): {c_final}")
```

```python
#Plot the line from iter 1 to iter 10(in dotted line)


import numpy as np
import matplotlib.pyplot as plt

def plot_gradient_steps(x, y):
    m = 0
    c = 0
    iterations = 10
    n = len(x)
    learning_rate = 0.01
    
    # Set up the plot for the data points
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='red', marker='+', s=100, label='Original Data')
    
    for i in range(iterations):
        y_predicted = m * x + c
        
        # Plot the line for the current iteration
        # We use a dotted line and vary the color/transparency
        plt.plot(x, y_predicted, linestyle='--', alpha=0.6, label=f'Iter {i+1}')
        
        # Gradient descent calculation
        dm = (-2/n) * sum(x * (y - y_predicted))
        dc = (-2/n) * sum(y - y_predicted)
        
        m = m - learning_rate * dm
        c = c - learning_rate * dc

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('First 10 Iterations of Gradient Descent')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.tight_layout()
    plt.show()

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

plot_gradient_steps(x, y)
```
